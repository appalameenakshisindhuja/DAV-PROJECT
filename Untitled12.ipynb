```python
import pandas as pd
df = pd.read_csv("./data/blinkit_customers.csv")
df.head()
```

```python
# prompt: Using dataframe df: generate a bar graph for different color bars using matplotlib

import matplotlib.pyplot as plt

# Assuming 'customer_segment' is the categorical variable and 'total_orders' is the numerical value
customer_segments = df['customer_segment'].unique()
total_orders = df.groupby('customer_segment')['total_orders'].sum()

# Creating the bar graph
plt.figure(figsize=(8, 6))
plt.bar(customer_segments, total_orders, color=['skyblue', 'lightcoral', 'lightgreen', 'lightblue']) # Customize colors here

# Customize the chart elements
plt.xlabel("Customer Segment")
plt.ylabel("Total Orders")
plt.title("Total Orders by Customer Segment")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping

# Display the chart
plt.show()
```

```python


import matplotlib.pyplot as plt
import numpy as np


values = df['sales']

# Create the polar plot
angles = np.linspace(0, 2 * np.pi, len(values), endpoint=False)
values = np.concatenate((values, [values[0]]))  # Close the circle
angles = np.concatenate((angles, [angles[0]]))

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(angles, values)
ax.fill(angles, values, alpha=0.25)
ax.set_title("Polar Plot of Sales") # Replace with 'promised_time' if it is an existing column name
plt.show()
```

SURFACE PLOT


```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np


data = {
    'product_id': np.tile([101, 102, 103, 104], 4),  # 4 products repeated
    'date': np.repeat(pd.date_range(start='2024-01-01', periods=4), 4),
    'stock_received': np.random.randint(10, 100, 16)  # Random stock data
}
df = pd.DataFrame(data)

df = df.sort_values(by=['product_id', 'date'])


df_pivot = df.pivot(index='date', columns='product_id', values='stock_received')

x = df_pivot.columns  # Unique product IDs
y = df_pivot.index  # Unique dates
z = df_pivot.values  # Reshaped stock_received

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Customize the layout
fig.update_layout(
    title='3D Surface Plot of Stock Received',
    autosize=False, width=800, height=800,
    margin=dict(l=65, r=50, b=65, t=90),
    scene=dict(xaxis_title='Product ID', yaxis_title='Date', zaxis_title='Stock Received')
)

fig.show()
```
